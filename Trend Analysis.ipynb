```python
import pandas as pd
import matplotlib.pyplot as plt
xls = pd.ExcelFile('Desktop/CloudTara/Apple_Train.xlsx')
df = pd.read_excel(xls, 'Quarterly')
print(df)
```

```python
df['Revenue1'] = df['TTM Revenue'].str.replace('$','')
```

```python
df['Revenue'] = df['Revenue1'].str.replace('B','')
```

```python
df.drop(['Revenue1'], axis=1)
```

```python
df['a'] = df['TTM EBITDA'].str.replace('$','')
```

```python
df['EBITDA'] = df['a'].str.replace('B','')
```

```python
df.drop(['Revenue1','a'], axis=1)
```

```python
df["Revenue"] = pd.to_numeric(df["Revenue"])
df["EBITDA"] = pd.to_numeric(df["EBITDA"])
```

```python
df['Quarterly Growth %'] = ((df['Revenue'] - df['Revenue'].shift(1))/df['Revenue'].shift(1))*100
```

```python
df.drop(['Revenue1','a'], axis=1)
```

```python
plt.plot(df.index, df['Quarterly Growth %'])
```

```python
df['Quarterly Growth %'] = df['Quarterly Growth %'].fillna(0)
```

```python
df['Valuation Date']=pd.to_datetime(df['Valuation Date'])
df=df.set_index('Valuation Date')
```

```python
plt.rcParams['figure.figsize']=[12.0,8.0]
from datetime import datetime
start_date=datetime(2006, 12, 31)
end_date=datetime(2020, 3, 31)
df[(start_date <= df.index) & (df.index<=end_date)].plot(grid=True)
```

```python
import statsmodels.api as sm
decomposition=sm.tsa.seasonal_decompose(df['Quarterly Growth %'], model='additive')
fig=decomposition.plot()
plt.rcParams['figure.figsize']=[12.0,8.0]
```

```python
import matplotlib.dates as mdates
fig, ax = plt.subplots()
ax.grid(True)
year = mdates.YearLocator(month=12)
month = mdates.MonthLocator(interval=3)
year_format = mdates.DateFormatter('%Y')
month_format = mdates.DateFormatter('%m')
ax.xaxis.set_minor_locator(month)
ax.xaxis.grid(True, which = 'minor')
ax.xaxis.set_major_locator(year)
ax.xaxis.set_major_formatter(year_format)
plt.plot(df.index, df['Quarterly Growth %'], c='blue')
plt.plot(decomposition.trend.index, decomposition.trend, c='red')
```
